```python
x = new.drop(['Class'],axis='columns').values
y = new.Class
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,stratify=y)
```

```python
print(x_train.shape,x_test.shape)
```

```python
model_params = {
    
    'logistic_reg':{
        'model': LogisticRegression(),
        'params':{
            'C':[1,5,10,15,20]
        }
    },
    'randomeforest':{
        'model': RandomForestClassifier(),
        'params':{
            'n_estimators':[5,10,15,20,25]
        }
    }
}
```

```python
score = []
for mod_name,mod in model_params.items():
   clf = GridSearchCV(mod['model'], mod['params'], cv=5, return_train_score=False)
   clf.fit(x_train, y_train)
   score.append({'model': mod_name, 'best_score': clf.best_score_, 'best_params':clf.best_params_})
```

```python
score
```

```python
model = LogisticsRegression(C=10).fit(x_train,y_train)
```

```python
model.score(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
matrix = confusion_matrix(y_test,y_pred)
matrix
```

```python
plt.figure(figsize=5,5)
sns.heatmap(matrix,annot=True,fmt='.2f',cmap='Blues')
plt.xlabel('Prediction')
plt.ylabel('Actuals')
plt.show()
```
